```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from RLU import RLUnivariada
```

```python
dataset = pd.read_csv('testeRLU.csv') #https://www.kaggle.com/code/erkanhatipoglu/univariate-linear-regression-from-scratch/notebook
```

```python
#gráfico de dispersão
plt.scatter(dataset['x'],dataset['y'], marker= 'x', c='r' )
plt.title("Input x Output")
plt.ylabel("Outuput")
plt.xlabel("Input")
plt.show()
```

```python
modelo =  RLUnivariada()
```

```python
#Treinando modelo
x = dataset['x']
target = dataset['y']
modelo.treino(x,target)
```

```python
predicoes =  modelo.predicao(x)
```

```python
plt.scatter(x,target, marker= 'x', c='r', label = "Valores reais" )
plt.plot(x,predicoes,c ='b', label= "Minha Predição")
plt.title("Input x Output")
plt.ylabel("Outuput")
plt.xlabel("Input")
plt.legend()
plt.show()
```

```python
#Valores de w e b com função Custo minimizada
print(f'w= {modelo.w}')
print(f'b= {modelo.b}')
print(f'J(w,b) = {modelo.fCusto(x,target)}')
```
