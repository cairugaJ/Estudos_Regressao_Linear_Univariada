## Entendendo Cost Function (Função de Custo / Função de Perda)
Treino: simulação simples de  regressão linear univariada e avaliação dos parâmetros w e b

curso: https://www.coursera.org/learn/machine-learning/home/week/1

```python
#Import de bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_treino = np.array([1.0,2.0,3.0]) #variável de entrada
y_treino = np.array([1.0,2.0,3.0]) #variável target
```

### Teste inicial incluindo apenas o parâmetro w

```python
valores_w =  np.arange(start= -.5, stop=3, step=0.5) #valores de w para testar qual minimiza J(w)
```

```python
def fw (w,x):
    '''Função que calcula valores estimados para y'''
    m = x.shape
    f_w = np.zeros(m)
    for i in range(len(x)):
        f_w[i] = w*x[i]
    return f_w    
```

```python
def Jw (y_hat,y):
    soma = 0
    '''Função que calcula Squared Error Cost Function'''
    for i in range(len(y)):
        soma+= (y_hat[i] - y[i])**2
    j = soma/2 *len(y)
    return j         
```

```python
lEst = []  #lista que armazenará as listas com valores estimados para cada valor de w
for valor in valores_w:
     est = fw(valor,x_treino)
     lEst.append(est)
```

```python
lJ = [] #Lista que armazenará valores da função perda com w diferentes
for est in lEst:
    lJ.append(Jw(est,y_treino))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].scatter(x_treino, y_treino, marker='x', c='r', label="Valor real")
for i in range(len(valores_w)):     
    axes[0].plot(x_treino, lEst[i], label=f"w = {valores_w[i]}")  
axes[0].set_title("Predições x valores reais") 
axes[0].set_ylabel("Output") 
axes[0].set_xlabel("Input") 
axes[0].legend()

axes[1].plot(valores_w,lJ, c='b')
axes[1].set_title("Função de Perda (J) em relação a w") 
axes[1].set_ylabel("J") 
axes[1].set_xlabel("Valor de w")
j_min = np.min(lJ) 
w_min = valores_w[np.argmin(lJ)] 
axes[1].scatter(w_min, j_min, c='r')  
```
