```python
#Import de bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
#x_treino é a variável de entrada (área da casa em metros quadrados)
#y_treino é o alvo/target (preço da casa em reais)
x_treino = np.array([100,200])
y_treino = np.array([30000,50000])
```

```python
#número de exemplos de treino "m"
m = len(x_treino)
print(f"O número de exemplos de treino é {m}")
```

```python
#Selecionando exemplo de treino pelo seu índice
i = 0 #pegar primeiro exemplo
x_i = x_treino[i]
y_i = y_treino[i]
print(f"(x^({i}),y^({i})) = ({x_i},{y_i})")
```

### Plotando os dados

```python
#gráfico de dispersão
plt.scatter(x_treino,y_treino, marker= 'x', c='r' )
#título
plt.title("Preço de Casas")
plt.ylabel("Preço em reais")
plt.xlabel("Área em metros quadrados")
plt.show()
```

### Definindo Função / Modelo manualmente
f_w,b(x) = wx + b

```python
w = 100
b = 100
print(f"w = {w}")
print(f"b = {b}")
```

```python
def f (w,b,x):
    '''Função que calcula valores estimados para y'''
    m = x.shape
    f_wb = np.zeros(m)
    for i in range(len(x)):
        f_wb[i] = w*x[i] + b
    return f_wb    
```

```python
estimativas = f(w,b,x_treino)
plt.plot(x_treino,estimativas,c ='b', label= "Minha Predição")
plt.scatter(x_treino,y_treino, marker= 'x', c='r', label = "Valores reais" )
plt.title("Preço de Casas")
plt.ylabel("Preço em reais")
plt.xlabel("Área em metros quadrados")
plt.legend()
plt.show()
```

```python
#Testando outros valores para w e b (tentar melhorar o modelo)
w = 200
b = 10000
estimativas_atualizadas = f(w,b,x_treino)
plt.plot(x_treino,estimativas_atualizadas,c ='b', label= "Minha Predição")
plt.scatter(x_treino,y_treino, marker= 'x', c='r', label = "Valores reais" )
plt.title("Preço de Casas")
plt.ylabel("Preço em reais")
plt.xlabel("Área em metros quadrados")
plt.legend()
plt.show()
```

```python
#Estimando valor de casa de 300 metros quadrados
valor_estimado = w*300 + b
print(f"Uma casa de 300 metros quadrados tem preço estimado de {valor_estimado} reais")
```
